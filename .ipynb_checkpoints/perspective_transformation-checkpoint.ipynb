```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
# define image size
height = 6
width = 8
# define laser pointer intrinsics
f = 0.5 # mm, focal distance
alpha = 1 # aspect ratio
fx = f*alpha
fy = f*alpha

cx = 0.5*width # px, principle point in x direciton
cy = 0.5* height # py, principle point in y direction

A = np.array([[fx,0,cx],[0,fy,cy],[0,0,1]])

print (A)
```

```python
# define lens pattern
o = [0,0,0] # origin
p1 = [-1,2]
p2 = [0,-1]
p3 = [4,3]
p4 = [-3,-3]

# [[x,y,z],[x,y,z]....]
# define index
x_ind = 0
y_ind = 1
z_ind = 2
p = np.vstack((p1,p2,p3,p4))

print(p)
print (p[:,0])

# plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(p[:,x_ind],p[:,y_ind],f,color='blue')
ax.scatter(o[x_ind],o[y_ind],o[z_ind],color='green')
ax.set_xlabel('X_c')
ax.set_ylabel('Y_c')
ax.set_zlabel('Z_c')

plt.show()
```

```python
# define the ceiling plane

X = np.arange(-40, 50, 80)
Y = np.arange(-30, 40, 60)
X, Y = np.meshgrid(X, Y)
D = 5
Z = D

# plot
fig = plt.figure()

ax = fig.gca(projection='3d')
ax.plot_surface(X,Y,Z,cmap=cm.cool,alpha=0.5)
ax.set_xlabel('X_c')
ax.set_ylabel('Y_c')
ax.set_zlabel('Z_c')

plt.show()
```

```python
# solve for each intersection point
p_proj = np.ndarray((4,3))

# plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in range (0,p.shape[0]):
    t = D/f
    p_proj[i,x_ind] = t*p[i,x_ind]
    p_proj[i,y_ind] = t*p[i,y_ind]
    p_proj[i,z_ind] = D
    print (p_proj[i,:])
    ax.plot([p_proj[i,x_ind],o[x_ind]],[p_proj[i,y_ind],o[x_ind]],[p_proj[i,z_ind],o[x_ind]],color='blue')

ax.scatter(o[x_ind],o[y_ind],o[z_ind],color='green')
ax.set_xlabel('X_c')
ax.set_ylabel('Y_c')
ax.set_zlabel('Z_c')

plt.show()
```

```python
# plot
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.scatter(p[:,x_ind],p[:,y_ind],f)
ax.scatter(o[x_ind],o[y_ind],o[z_ind],color='green')
ax.plot_surface(X,Y,Z,cmap=cm.cool,alpha=0.5)
for i in range (0,p.shape[0]):
    ax.plot([p_proj[i,x_ind],o[x_ind]],[p_proj[i,y_ind],o[x_ind]],[p_proj[i,z_ind],o[x_ind]],color='blue',linestyle='dashed')
    ax.scatter(p_proj[i,x_ind],p_proj[i,y_ind],p_proj[i,z_ind],color='red')
ax.set_xlabel('X_c')
ax.set_ylabel('Y_c')
ax.set_zlabel('Z_c')

plt.show()
```

```python
# calculate transformation
# case 1, world coordinate is the same as camera
cv2.solvePnP(p_proj,p,A,distCoeffs= 0)
```
